```python
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Sample documents
documents = [
    "The quick brown fox jumps over the lazy dog.",
    "A cat and a dog are best friends.",
    "I love my pet cat.",
]

# Initialize CountVectorizer to convert text into bag of words
vectorizer = CountVectorizer()

# Fit and transform the documents to obtain the bag of words representation
bow_matrix = vectorizer.fit_transform(documents)

# Get the list of words in the vocabulary
vocabulary = vectorizer.get_feature_names_out()

# Print the vocabulary
print("Vocabulary:", vocabulary)

# Visualize the bag of words using a word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(dict(zip(vocabulary, bow_matrix.toarray().sum(axis=0))))
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud - Bag of Words")
plt.show()

# Create a table of words with their frequencies
word_frequencies = bow_matrix.toarray().sum(axis=0)
word_table = {word: freq for word, freq in zip(vocabulary, word_frequencies)}
print("\nTable of Words with Frequencies:")
for word, freq in word_table.items():
    print(f"{word}: {freq}")
```

```python
king=nlp(u"king").vector
man=nlp(u"man").vector
woman=nlp(u"woman").vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Example vectors
vector1 = [[1, 2, 3]]
vector2 = [[4, 5, 6]]

# Calculate cosine similarity
similarity = cosine_similarity(vector1, vector2)

print("Cosine Similarity:", similarity)
```

```python
# Calculate the new vector representing "queen" by analogy: king - man + woman
new_vector = king - man + woman

# Initialize a list to store computed vectors
computed_vectors = []

# Iterate over all words in the vocabulary
for word in nlp.vocab:
    # Check if the word has a vector representation
    if word.has_vector:
        # Calculate the cosine similarity between the new vector and the word's vector
        similarity = cosine_similarity(np.array([new_vector]), np.array([word.vector]))[0][0]
        # Append the word and its similarity to the list of computed vectors
        computed_vectors.append((word, similarity))

# Sort the computed vectors by similarity in descending order
computed_vectors = sorted(computed_vectors, key=lambda item: -item[1])

# Print the top 10 words based on similarity
print([w[0].text for w in computed_vectors[:10]])
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Create a SentimentIntensityAnalyzer object
sid = SentimentIntensityAnalyzer()

# Analyze the sentiment of a sentence
sentence = "This movie is great!"
sentiment_scores = sid.polarity_scores(sentence)

# Print the sentiment scores
print(sentiment_scores)
```
